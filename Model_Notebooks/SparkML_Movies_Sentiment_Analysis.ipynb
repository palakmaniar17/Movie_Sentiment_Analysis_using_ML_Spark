# text cleaning

```python
from pyspark.sql import functions as fun
from pyspark.ml.feature import Tokenizer
from pyspark.ml.feature import StopWordsRemover
from pyspark.sql import types as spark_types
```

```python
# convert string to lower
cleaned_reviews =  raw_spark_df.withColumn("cleaned_text", fun.lower("review"))
```

```python
cleaned_reviews.show(5)
```

```python
# remove numbers, punctuations and white spaces from text
cleaned_reviews = cleaned_reviews.withColumn("cleaned_text", fun.trim(fun.regexp_replace(fun.col("cleaned_text"), "[0-9\\p{Punct}]", "")))
```

```python
cleaned_reviews.show(5)
```

```python
# tokenize text
tokenizer = Tokenizer(inputCol="cleaned_text", outputCol="words")
```

```python
tokenized_reviews_df = tokenizer.transform(cleaned_reviews)
```

```python
tokenized_reviews_df.show(5)
```

```python
# Remove stop words
stopword_remover = StopWordsRemover(inputCol="words", outputCol="review_wo_stopwords")
```

```python
reviews_wo_stopwords_df = stopword_remover.transform(tokenized_reviews_df)
```

```python
reviews_wo_stopwords_df.show()
```

```python
# top words
top_words_df = reviews_wo_stopwords_df.select("review_wo_stopwords").withColumn("words", fun.explode("review_wo_stopwords"))
```

```python
# show top 10 words
top_words_df.groupBy("review_wo_stopwords").count().orderBy(fun.desc("count")).limit(10).show(10)
```

```python
# define user defined function to remove blank spaces
def remove_blank_spaces(words):
  return [word for word in words if word.strip()]
```

```python
# define spark udf
remove_blanc_spaces_udf = fun.udf(remove_blank_spaces, spark_types.ArrayType(spark_types.StringType()))
```

```python
# final cleaned dataframe
cleaned_df = reviews_wo_stopwords_df.withColumn("cleaned_words",
                                   remove_blanc_spaces_udf(fun.col("review_wo_stopwords")))
```

```python
cleaned_df.select("cleaned_words").show(5, truncate=False)
```

# modelling

```python
from pyspark.ml.feature import Tokenizer, HashingTF, IDF
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
```

```python
RANDOM_STATE = 42
```

```python
# convert cleaned words in plain text
final_df = cleaned_df.withColumn("clean_reviews", fun.concat_ws(" ", "cleaned_words"))
```

```python
final_df.show(5)
```

```python
# tokenize and remove stop words
tokenizer = Tokenizer(inputCol="clean_reviews", outputCol="clean_words")
stopwords_remover = StopWordsRemover(inputCol=tokenizer.getOutputCol(), outputCol="filtered_words")
```

```python
# convert label
indexer = StringIndexer(inputCol="sentiment", outputCol="label")
```

```python
# tfidf features
hashingTF = HashingTF(inputCol=stopwords_remover.getOutputCol(), outputCol="raw_features", numFeatures=100)
idf = IDF(inputCol=hashingTF.getOutputCol(), outputCol="features")
```

## Logistic Regression

```python
# regression model
lr = LogisticRegression(maxIter=10, regParam=0.01)
```

```python
# pipeline for feature extraction and model
pipeline = Pipeline(stages=[tokenizer, stopwords_remover, indexer, hashingTF, idf, lr])
```

```python
# split data in train-test
train_data, test_data = final_df.randomSplit([0.8, 0.2], seed=RANDOM_STATE)
```

```python
# model fit
start_time = time.time()
%time model = pipeline.fit(train_data)
end_time = time.time()
```

```python
execution_time = end_time - start_time
```

```python
# predict on test
predictions = model.transform(test_data)
```

```python
predictions.show(5)
```

```python
# evaluate
METRICS = ["f1", "weightedPrecision", "weightedRecall", "accuracy"]
SCORES = []
for metric in METRICS:
  evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName=metric)
  score = evaluator.evaluate(predictions)
  print(f"{metric} Score: {score}")
  SCORES.append(score)
```

```python
log_reg_data_dict = {'model':'logistic_regression',
             'train_time':execution_time,
             'f1_score':SCORES[0],
             'precision':SCORES[1],
             'recall':SCORES[2],
             'accuracy':SCORES[3]}
```

## SVM with parameter tuning

```python
from pyspark.ml.classification import LinearSVC
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
```

```python
# svm model
svm = LinearSVC()
```

```python
# svm pipeline
svm_pipeline = Pipeline(stages=[tokenizer, stopwords_remover, indexer, hashingTF, idf, svm])
```

```python
# define grid
param_grid_svm = ParamGridBuilder() \
    .addGrid(svm.regParam, [0.01, 0.1, 1.0]) \
    .addGrid(svm.maxIter, [10]) \
    .build()
```

```python
# cross validate and improve f1 score
svm_crossval = CrossValidator(estimator=svm_pipeline,
                              estimatorParamMaps=param_grid_svm,
                             evaluator=MulticlassClassificationEvaluator(labelCol="label",
                                                                         predictionCol="prediction",
                                                                         metricName="f1"),
                              numFolds=3)
```

```python
# Fit the SVM model
start_time = time.time()
svm_cv_model = svm_crossval.fit(train_data)
end_time = time.time()

execution_time = end_time - start_time
```

```python
# Make predictions on the test data
svm_test_predictions = svm_cv_model.transform(test_data)
```

```python
# evaluate
METRICS = ["f1", "weightedPrecision", "weightedRecall", "accuracy"]
SCORES = []
for metric in METRICS:
  evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName=metric)
  score = evaluator.evaluate(svm_test_predictions)
  print(f"{metric} Score: {score}")
  SCORES.append(score)
```

```python
svm_data_dict = {'model':'svm_model',
             'train_time':execution_time,
             'f1_score':SCORES[0],
             'precision':SCORES[1],
             'recall':SCORES[2],
             'accuracy':SCORES[3]}
```

## RandomForest Model

```python
from pyspark.ml.classification import RandomForestClassifier
```

```python
# RandomForestClassifier
rf = RandomForestClassifier(featuresCol="features", labelCol="label", numTrees=100, seed=42)
```

```python
# random forest pipeline
rf_pipeline = Pipeline(stages=[tokenizer, stopwords_remover, indexer, hashingTF, idf, rf])
```

```python
param_grid_rf = ParamGridBuilder() \
    .addGrid(rf.maxDepth, [5, 10]).build()
```

```python
# CrossValidator for RandomForest
crossval_rf = CrossValidator(estimator=rf_pipeline,
                             estimatorParamMaps=param_grid_rf,
                             evaluator=MulticlassClassificationEvaluator(labelCol="label",
                                                                         predictionCol="prediction",
                                                                         metricName="f1"))
```

```python
# Fit the SVM model
start_time = time.time()
model_rf = crossval_rf.fit(train_data)
end_time = time.time()

execution_time = end_time - start_time
```

```python
# Make predictions on the test data
predictions = model_rf.transform(test_data)
```

```python
# evaluate
METRICS = ["f1", "weightedPrecision", "weightedRecall", "accuracy"]
SCORES = []
for metric in METRICS:
  evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName=metric)
  score = evaluator.evaluate(svm_test_predictions)
  print(f"{metric} Score: {score}")
  SCORES.append(score)
```

```python
rf_data_dict = {'model':'random_forest',
             'train_time':execution_time,
             'f1_score':SCORES[0],
             'precision':SCORES[1],
             'recall':SCORES[2],
             'accuracy':SCORES[3]}
```

```python
# metrics df
metrics_df = pd.DataFrame([log_reg_data_dict, rf_data_dict, svm_data_dict])
```
